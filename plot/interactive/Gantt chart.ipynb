```python
# Importing the matplotlb.pyplot 
import matplotlib.pyplot as plt 
  
plt.style.use('ggplot')

# Declaring a figure "gnt" 
fig, ax = plt.subplots(figsize=(12, 8)) 
  
# Setting Y-axis limits 
ax.set_ylim(0, 50) 
  
# Setting X-axis limits 
ax.set_xlim(0, 140) 
  
# Setting labels for x-axis and y-axis 
# ax.set_xlabel('seconds since start') 
# ax.set_ylabel('Processor') 
  
# Setting ticks on y-axis 
ax.set_yticks([15, 25, 35, 45]) 
# Labelling tickes of y-axis 
ax.set_yticklabels(['Thesis Writing', 'Benchmarking', 'ChipYard Int.','Textual Front-end']) 
  
# Setting graph attribute 
# ax.grid(True)
ax.xaxis.grid()
ax.set_axisbelow(True)
  
#Set xlabel
ax.set_xticklabels(['Jan', 'Feb', 'March', 'April', 'May', 'Jun', 'July', 'Aug'])
# ax.set_xlabel('Timeline',fontsize=24)


# Declaring a bar in schedule 
ax.broken_barh([(60, 20)], (20, 9), facecolors =('tab:orange'),alpha=1) 
  
# Declaring multiple bars in at same level and same width 
ax.broken_barh([(20, 20)], (40, 9), 
                         facecolors ='tab:blue') 

ax.broken_barh([(40, 20)], (30, 9), facecolors =('tab:red'),alpha=1) 

ax.broken_barh([(80, 40)], (10, 9), facecolors =('tab:green'),alpha=1)

plt.tick_params(axis='both', which='major', labelsize=20)
  
# plt.savefig("gantt1.png") 
# plt.show()
fig.savefig('test.pdf',facecolor=fig.get_facecolor(), bbox_inches='tight')
```
