```python
## Importing the matplotlb.pyplot 
import matplotlib.pyplot as plt 
  
plt.style.use('ggplot')

# Declaring a figure "gnt" 
fig, ax = plt.subplots(figsize=(12, 8)) 
  
# Setting Y-axis limits 
ax.set_ylim(0, 50) 
  
# Setting X-axis limits 
ax.set_xlim(0, 140) 
  
# Setting labels for x-axis and y-axis 
# ax.set_xlabel('seconds since start') 
# ax.set_ylabel('Processor') 
  
# Setting ticks on y-axis 
ax.set_yticks([15, 25, 35, 45]) 
# Labelling tickes of y-axis 
ax.set_yticklabels(['Thesis Writing', 'Benchmarking', 'ChipYard Int.','Frontend/Serialization']) 
  
# Setting graph attribute 
# ax.grid(True)
# ax.xaxis.grid()
ax.yaxis.grid()
ax.set_axisbelow(True)
  
#Set xlabel
ax.set_xticklabels(['Jan', 'Feb', 'March', 'April', 'May', 'Jun', 'July', 'Aug'])
# ax.set_xlabel('Timeline',fontsize=24)

  
# Declaring multiple bars in at same level and same width 
ax.broken_barh([(20, 20)], (40, 9), 
                         facecolors ='steelblue')
ax.broken_barh([(40, 20)], (40, 9), 
                         facecolors ='skyblue') 

chipyard_imp  = ax.broken_barh([(40, 20)], (30, 9), facecolors =('firebrick')) 
chipyard_test = ax.broken_barh([(60, 20)], (30, 9), facecolors =('salmon'))

# Declaring a bar in schedule 
ax.broken_barh([(60, 20)], (20, 9), facecolors =('sandybrown')) 


ax.broken_barh([(80, 40)], (10, 9), facecolors =('forestgreen'))

plt.tick_params(axis='both', which='major', labelsize=22)

ax.legend(['Implementation', 'Testing', 'Implementation', 'Testing'], fontsize = 20, ncol=2, loc='lower center')
# ax.legend([chipyard_imp, chipyard_test], ['ChipYard Imp.', 'ChipYard Testing'], fontsize = 20, ncol = 2, loc='lower center')

# plt.savefig("gantt1.png") 
# plt.show()
fig.savefig('test.pdf',facecolor=fig.get_facecolor(), bbox_inches='tight')
```
